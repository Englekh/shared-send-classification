```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import datetime
import time
import os
```

## General data disrciption

```python
df = pd.read_csv('res.csv')

BLOCK_SIZE = 2016 

# Grouping data by blocks
df['group'] = df['num'] // BLOCK_SIZE
df['date_m'] = df['year'].to_numpy() + (df['month'] - 1) / 12 + (df['day'] - 1) / 365
```

```python
labels = ['Separable', 'Ambigeous', 'Simple', 'Time Limit']
sizes = [df['VERDICT SEPARABLE'].sum(), df['VERDICT AMBIGUOUS'].sum(), df['VERDICT SIMPLE'].sum(), df['TIME LIMIT EXTENDED'].sum()]

fig1, ax1 = plt.subplots()
plt.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Transaction types')
plt.show()
```

```python
keys = ['VERDICT SEPARABLE', 'VERDICT AMBIGUOUS', 'VERDICT SIMPLE', 'TIME LIMIT EXTENDED']
group_s = df.groupby(['group']).sum().reset_index()
group_m = df.groupby(['group']).mean().reset_index()

plt.figure(figsize=(12, 7))

for i in range(4):        
    plt.plot(group_m['date_m'], group_s[keys[i]], label = labels[i])

plt.yscale('log')
plt.xlabel('Year', fontsize = 20)
plt.ylabel('Transactions log', fontsize = 20)
plt.title('Transactions num', fontsize = 24)
plt.legend(prop={'size': 12}, loc='upper left')
plt.show()
```

## Time limit data description

```python
df_t = pd.read_csv('times.csv')
cols = ['5', '10', '30', '60', '300']
```

```python
labels = ['Separable', 'Ambigeous', 'Time Limit', 'Simple']
strings = ['VERDICT SEPARABLE', 'VERDICT AMBIGUOUS', 'TIME LIMIT EXTENDED', 'SIMPLE']
values = {}

for i in range(4):
    values[labels[i]] = []
    for col in cols:
        values[labels[i]].append(np.sum(df_t[col].map(lambda x: strings[i] in x)) / 1000)
    values[labels[i]] = np.array(values[labels[i]])
        
plt.figure(figsize=(7, 7))
    
plt.plot(cols, values[labels[0]] , label = labels[0])
plt.fill_between(cols, values[labels[0]], 0)
plt.plot(cols, values[labels[0]] + values[labels[1]], label = labels[1])
plt.fill_between(cols, values[labels[0]], values[labels[0]] + values[labels[1]])
plt.plot(cols, values[labels[0]] + values[labels[1]] + values[labels[2]], label = labels[2])
plt.fill_between(cols, values[labels[0]] + values[labels[1]], values[labels[0]] + values[labels[1]] + values[labels[2]])
plt.plot(cols, values[labels[0]] + values[labels[1]] + values[labels[2]], label = labels[2])
plt.fill_between(cols, values[labels[0]] + values[labels[1]], values[labels[0]] + values[labels[1]] + values[labels[2]])

plt.xlabel('Seconds', fontsize = 20)
#plt.xscale('log')
plt.ylabel('Transactions log', fontsize = 20)
plt.title('Transactions num', fontsize = 24)
plt.legend(prop={'size': 12}, loc='upper left')
plt.show()
```

```python
res = []
for col in cols:
    labels = ['Separable', 'Ambigeous', 'Time Limit', 'Simple']
    strings = ['VERDICT SEPARABLE', 'VERDICT AMBIGUOUS', 'TIME LIMIT EXTENDED', 'SIMP']
    sizes = []
    
    for el in strings:
        sizes.append(np.sum(df_t[col].map(lambda x: el in x)))
    res.append(sizes)
    
    fig1, ax1 = plt.subplots()
    ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
            shadow=True, startangle=90)
    ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.title('Transactions ' + str(col) + ' sec' )
    plt.show()
```
